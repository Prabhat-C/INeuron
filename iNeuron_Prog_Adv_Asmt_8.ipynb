1. Given a sentence as txt, return True if any two adjacent words have this property: One word ends with a vowel, while the word immediately after begins with a vowel (a e i o u).

Examples

vowel_links("a very large appliance") ➞ True

vowel_links("go to edabit") ➞ True

vowel_links("an open fire") ➞ False

vowel_links("a sudden applause") ➞ False


```python
import re

def vowel_links(string):
    vowels = ['a','e','i','o','u']
    spaces = re.finditer(" ", string)    #store all the matches in the iterator
    for i in spaces:
        before = i.span()[0]-1          #indexes before and after the space
        after = i.span()[1]        
        
        if (string[before] in vowels) and (string[after] in vowels):   #checking characters at before and after 
            return True
    
    else:
        return False 
    
    
```

3. Create a function that returns the characters from a list or string r on odd or even positions, depending on the specifier s. The specifier will be "odd" for items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2, 4, 6, ...).

Examples

char_at_pos([2, 4, 6, 8, 10], "even") ➞ [4, 8]
 4 & 8 occupy the 2nd & 4th positions

char_at_pos("EDABIT", "odd") ➞ "EAI"
 "E", "A" and "I" occupy the 1st, 3rd and 5th positions

char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd") ➞ ["A", "B", "T", "A", "I", "Y"]


```python
def char_at_pos(lis, spec):
    result = []
    if spec == 'even':
        for i,x in enumerate(lis):
            if i%2!=0:
                result.append(lis[i])
    if spec == 'odd':
        for i,x in enumerate(lis):
            if i%2==0:
                result.append(lis[i])
                
    if type(lis)==list:            
        return result
    else:
        return ''.join(result)
```

4. Write a function that returns the greatest common divisor of all list elements. If the greatest common divisor is 1, return 1.

Examples

GCD([10, 20, 40]) ➞ 10

GCD([1, 2, 3, 100]) ➞ 1

GCD([1024, 192, 2048, 512]) ➞ 64


```python
from functools import reduce 

def fact(n):      #function to find factors of a num
    factors = []
    i=2
    while i<=n:
        if n%i==0:
            factors.append(i)
            n=n/i
        else:
             i+=1              
    return factors


def GCD(lis):             #function to find GCD
    factors_all = []      #list to hold list of factors 
    for i in lis:
        factors_all.append(fact(i))
    
    common = []
    l = len(factors_all)
    
    for i in factors_all[0]:      #check the presence of first list elements in other lists 
        for j in range(1,l):
            if i in factors_all[j]:
                if j==l-1:            #if element is present in all lists, then only j will reach l-1
                    factors_all[j].remove(i)   #remove the element from last list, it will not be appended again            
            else:
                break
        else:                    #if loop runs without break, add the element as a factor 
            common.append(i)          
            
    if len(common)==0:
        return 1
    else: 
        return reduce(lambda x,y: x*y, common)  # to return multiplication use reduce on the list of common factors 
```

5. A number/string is a palindrome if the digits/characters are the same when read both forward and backward. Examples include "racecar" and 12321. Given a positive number n, check if n or the binary representation of n is palindromic. Return the following:

- "Decimal only." if only n is a palindrome.
- "Binary only." if only the binary representation of n is a palindrome.
- "Decimal and binary." if both are palindromes.
- "Neither!" if neither are palindromes.

Examples

palindrome_type(1306031) ➞ "Decimal only."
 decimal = 1306031
 binary  = "100111110110110101111"

palindrome_type(427787) ➞ "Binary only."
 decimal = 427787
 binary  = "1101000011100001011"

palindrome_type(313) ➞ "Decimal and binary."
 decimal = 313
 binary  = 100111001

palindrome_type(934) ➞ "Neither!"
 decimal = 934
 binary  = "1110100110"


```python
def palindrome_type(num):
    
    d,b=0,0
    
    if list(str(num))== list(reversed(str(num))):
        d=1
    if list(str(bin(num)[2:])) == list(reversed(str(bin(num)[2:]))):
        b=1
        
    if d==0 and b==0:
        return "Neither!"
    if d==1 and b==0:
        return "Decimal only."
    if d==0 and b==1:
        return "Binary only."
    if d==1 and b==1:
        return "Decimal and binary."
        
        
```
